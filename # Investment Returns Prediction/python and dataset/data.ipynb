```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv("Finance_Trends.csv")  # change filename to your actual file
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['age'], bins=15, kde=True, color='skyblue')
plt.title("Age Distribution of Respondents")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset


# Create the grouped bar chart
plt.figure(figsize=(10,6))
sns.countplot(x='What are your savings objectives?', hue='gender', data=df, palette='Set2')

# Add titles and labels
plt.title("Savings Objectives vs Gender", fontsize=14)
plt.xlabel("Savings Objectives", fontsize=12)
plt.ylabel("Number of Respondents", fontsize=12)
plt.xticks(rotation=25)
plt.legend(title='Gender')
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.scatterplot(x='Mutual_Funds', y='Equity_Market', hue='gender', data=df)
plt.title("Mutual Funds vs Equity Market")

plt.subplot(1,2,2)
sns.scatterplot(x='Gold', y='Fixed_Deposits', hue='gender', data=df)
plt.title("Gold vs Fixed Deposits")

plt.tight_layout()
plt.show()
```

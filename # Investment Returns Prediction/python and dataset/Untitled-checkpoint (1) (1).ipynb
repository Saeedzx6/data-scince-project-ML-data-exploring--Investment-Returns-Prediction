```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Finance_Trends.csv')
df
```

```python
df = df.drop(['Reason_Equity','Reason_Mutual','Reason_Bonds','Reason_FD','Investment_Avenues'], axis=1)
df
```

```python

print("Missing values per column:")
print(df.isnull().sum())
```

```python
for col in df.columns:
    if df[col].dtype == 'object':
        
        df[col].fillna('Unknown', inplace=True)
    else:
        
        df[col].fillna(df[col].mean(), inplace=True)

print("\nMissing values after filling:")
print(df.isnull().sum())
```

```python
# Initial inspection
print("Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())
print("\nFirst 3 rows:")
display(df.head(3))
```

```python
df['age'] = pd.to_numeric(df['age'], errors='coerce')
print(f"Before filtering: {df['age'].describe()}")

df = df[df['age'].between(18, 38)].copy()

# Fill rare missing with median
df['age'] = df['age'].fillna(df['age'].median()).astype(int)

print(f"After cleaning: {df['age'].describe()}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


df['age'] = pd.to_numeric(df['age'], errors='coerce')
df = df.dropna(subset=['age'])

sns.set(style="whitegrid")

plt.figure(figsize=(8,5))
sns.histplot(df['age'], bins=10, kde=True, color="skyblue", edgecolor="black")

plt.title("Age Distribution", fontsize=14)
plt.xlabel("Age")
plt.ylabel("Count")

plt.savefig("age_distribution.pdf", format='pdf')

plt.show()

print("✅ PDF saved as 'age_distribution.pdf'")
```

```python
# Create the figure with subplots
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.countplot(data=df, x='Factor', hue='gender')
plt.title('Investment Factor Preference by Gender')

plt.subplot(1, 2, 2)
sns.countplot(data=df, x='Purpose', hue='Factor')
plt.title('Investment Purpose vs Factor')
plt.xticks(rotation=45)

# Save as PDF
plt.tight_layout()
plt.savefig('investment_analysis.pdf', dpi=300, bbox_inches='tight')
plt.show()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(12,6))
obj_counts = df['What are your savings objectives?'].value_counts()
sns.barplot(x=obj_counts.values, y=obj_counts.index)
plt.title('Primary Savings Objectives', fontsize=14)
plt.xlabel('Count')
plt.ylabel('Savings Objectives')

plt.tight_layout()
plt.savefig("savings_objectives.pdf", format='pdf')

plt.show()

print("✅ PDF saved as 'savings_objectives.pdf'")
```

```python
expect_mapping = {
    '10%-20%': 15,
    '20%-30%': 25,
    '30%-40%': 35
}

df['Expect_num'] = df['Expect'].map(expect_mapping)
```

```python
features = [
    'age',
    'gender',
    'Duration',
    'Invest_Monitor'
]

X = df[features]
y = df['Expect_num']
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=1/3, random_state=42
)

print(len(X_train))
```

```python
from sklearn import svm
regr = svm.SVR(kernel='linear', C= 1.0) # Train the model
regr.fit(X_train, y_train)
```

```python
from sklearn.svm import SVR

reg = SVR(kernel='linear', C=1.0)
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)
```

```python
y_pred = reg.predict(X_test)

print("Predicted values:", y_pred[:5])
print("Actual values:", y_test.values[:5])
```

```python
results = pd.DataFrame({
    'Actual Expected Return': y_test.values,
    'Predicted Expected Return': y_pred
})

results = results.reset_index(drop=True)

print(results.head(10))
```
